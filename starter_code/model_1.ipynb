```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co', 'koi_fpflag_ec', 'koi_period', 'koi_period_err1',
                       'koi_period_err2', 'koi_time0bk', 'koi_time0bk_err1', 'koi_steff_err2', 'koi_srad', 'koi_srad_err1',
                       'koi_srad_err2', 'ra', 'dec', 'koi_kepmag']]
selected_features.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# Assign X (data) and y (target)
X = selected_features
y = df["koi_disposition"]
```

```python
# Split the data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder

# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)


# Transform the training and testing data using the X_scaler

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

# Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
from tensorflow.keras.utils import to_categorical
```

# Train the Model



```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(model, {'C': [1, 10], 'penalty': ["l1", "l2"]}, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
